```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
from cropimage import Cropper
from google.colab.patches import cv2_imshow
```

#Detection of face from image

```python
path='/content/test12.jpg'
test_image=cv2.imread(path)
testimg_togray=cv2.cvtColor(test_image,cv2.COLOR_BGR2GRAY)
plt.imshow(testimg_togray,cmap='gray')
```

```python
def convertToRGB(imge):
    return cv2.cvtColor(imge,cv2.COLOR_BGR2RGB)
```

```python
haar_cascade_face=cv2.CascadeClassifier('/content/haarcascade_frontalface_alt.xml')
```

```python
def imgcropfun(pth):
  cropper = Cropper()

  result = cropper.crop(pth)
  cv2_imshow(result)

  cv2.imwrite('cropped.jpg', result)
  return result
```

```python
faces_rects = haar_cascade_face.detectMultiScale(testimg_togray, scaleFactor = 1.2, minNeighbors = 5);
print(len(faces_rects))
flag=len(faces_rects)
t=0
if(flag>0):
  print("face detected,returned image with max feature crop")
  img=convertToRGB(testimg_togray)
  crpimg=imgcropfun(path)
else:
  print("retake image,face not detected")
  
```
